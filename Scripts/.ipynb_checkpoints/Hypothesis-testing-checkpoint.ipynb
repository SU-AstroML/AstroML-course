```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(1)
```

```python
muB, sigmaB = 100, 10
muS, sigmaS = 150, 12
hB = stats.norm(muB, sigmaB)
hS = stats.norm(muS, sigmaS)
a = 0.1
N = 1e6
x_c = 120 #a 2sigma cut away from the mean for hB
alpha = 0.024
bg = hB.rvs(N)
src = hS.rvs(N)
```

```python
#src_hist = plt.hist(src, bins=1000, normed=True, histtype='step', color='g', lw=3, label='source')
#bg_hist = plt.hist(bg, bins=1000, normed=True, histtype='step', color='r', lw=3, label='background')
#plt.legend()
pS = a*hS.pdf(src)
pB = (1-a)*hB.pdf(bg)
plt.plot(src, pS, '.g', label='source')
plt.plot(bg, pB, '.r', label='background')
plt.vlines(x=x_c, ymin=0, ymax=0.04, color='k', lw=2)
plt.ylim([0,0.04])
plt.legend()
plt.show()
spurious = [x for x in bg if x>=x_c]
missed = [x for x in src if x<=x_c]
n_spurious, n_missed = float(len(spurious)), float(len(missed))
n_src = N*a - n_missed + n_spurious
print '(Spurious source, Missed source, True source):', (n_spurious, n_missed, n_src)

print '(Type I error rate, Type II error rate): ', round(n_spurious/N, 3), round(n_missed/N, 3)
eta = round((N*a - n_missed)/(N*a), 3)
epsilon = round(n_spurious/n_src, 3)

print '(completeness, contamination):', (eta, epsilon)
```

```python
Xcs = np.linspace(muB, muB+4.*sigmaB, 25)
ROC_curve = np.zeros_like(Xcs)
for i in range(len(Xcs)):
    x_c = Xcs[i]
    spurious = [x for x in bg if x>=x_c]
    missed = [x for x in src if x<=x_c]
    n_spurious, n_missed = float(len(spurious)), float(len(missed))
    n_src = float(N*a - n_missed + n_spurious)
    eta = float((N*a - n_missed)/(N*a))
    epsilon = float(n_spurious/n_src)
    ROC_curve[i] = eta/epsilon
plt.semilogy(Xcs, ROC_curve, '-m')
```

## Figure 4.6 False discovery control (FDC) - Benjamini & Hochberg method

```python
# Set up the background and foreground distributions
background = norm(100, 10)
foreground = norm(150, 12)
f = 0.1

# Draw from the distribution
N = 1E6
X = np.random.random(N)
mask = (X < 0.1)
X[mask] = foreground.rvs(np.sum(mask))
X[~mask] = background.rvs(np.sum(~mask))

#------------------------------------------------------------
# Perform Benjamini-Hochberg method
p = 1 - background.cdf(X)
p_sorted = np.sort(p)

#------------------------------------------------------------
# plot the results
fig = plt.figure(figsize=(5, 3.75))
fig.subplots_adjust(bottom=0.15)
ax = plt.axes(xscale='log', yscale='log')

# only plot every 1000th; plotting all 1E6 takes too long
ax.plot(p_sorted[::1000], np.linspace(0, 1, 1000), '-k')
ax.plot(p_sorted[::1000], p_sorted[::1000], ':k', lw=1)

# plot the cutoffs for various values of expsilon
p_reg_over_eps = 10 ** np.linspace(-3, 0, 100)
for (i, epsilon) in enumerate([0.1, 0.01, 0.001, 0.0001]):
    x = p_reg_over_eps * epsilon
    y = p_reg_over_eps
    ax.plot(x, y, '--k')

    ax.text(x[1], y[1],
            r'$\epsilon = %.1g$' % epsilon,
            ha='center', va='bottom', rotation=70)

ax.xaxis.set_major_locator(plt.LogLocator(base=100))

ax.set_xlim(1E-12, 1)
ax.set_ylim(1E-3, 1)

ax.set_xlabel('$p = 1 - H_B(i)$')
ax.set_ylabel('normalized $C(p)$')

plt.show()
```
