## MLE

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop = stats.norm(0) #model
p = pop.pdf
x = pop.rvs(1000) #sample

#now we can calculate the likelihood of drawing just the first point
N_draws = 1
L = np.prod(p(x[0:N_draws])) #Product of the first N_draw elements
print 'L(1) = ', L

N_draws = 10
L = np.prod(p(x[0:N_draws]))
print 'L(10) = ',L

N_draws = 100
L = np.prod(p(x[0:N_draws]))
print 'L(100) = ', L

N_draws = 1000
L = np.prod(p(x[0:N_draws]))
Ls = np.sum(np.log(p(x[0:N_draws])))
print 'L(1000) = ',L
print 'ln(L(1000)) = ',Ls
```

```python
#lets define a data set drawn from a normal gaussian
N_data = 5
rv = stats.norm(0)
p = rv.pdf
x = rv.rvs(N_data)
mu = np.mean(x)
sigma = 1
print 'mu:', mu, 'sigma:', sigma
z = (x - mu) / sigma
Q = np.sum(z**2.0)
print 'Q:', Q

Q=np.zeros(1000)
for seed in range(0,1000):
    x = rv.rvs(N_data) #draw 100 points from this distribution
    mu = np.mean(x)
    z = (x - mu) / sigma
    Q[seed] = np.sum(z**2.0)

bins = np.arange(0,25,0.5)
result = np.histogram(Q,bins=bins)[0] #values of the histogram (each bin)
result = result/1000.0 #normalize the histogram

dof = N_data - 1 #remeber we fixed the sigma?
rv = stats.chi2(dof)
x = rv.rvs(100000)
result2 = np.histogram(x,bins=bins)[0]
result2 = result2/100000.0

plt.figure()
plt.bar(bins[:-1],result,width=0.5) #Qs
plt.plot(bins[:-1],result2, color='red', lw=5) #chi^2 dist
plt.show()
```

## Goodness of fit

```python
#first , let's generate the "data"
np.random.seed(1) #pass the random number generator a seed.

N  = 50 #get 50 data points
L0 = 10 #the intrinsic luminosity is 10 (data units)
dL = 0.2 #the error on each measurement is 0.2 (data units).

t = np.linspace(0, 1, N) #take each measurment over a time period [0,1] (time units)
L_obs = np.random.normal(L0, dL, N) #get data with a normal distribution around L0
```

```python
#now take a look at the "data"
plt.figure()
plt.plot([-0.1, 1.3], [L0, L0], ':k', lw=1) #true mean
plt.errorbar(t, L_obs, dL, fmt='.k', ecolor='gray', lw=1) #data
plt.xlim(-0.05, 1.05)
plt.ylim(8.6, 11.4)
plt.xlabel('time')
plt.ylabel('luminosity')
plt.show()
```

```python
# compute the mean and the chi^2/dof
mu = np.mean(L_obs) #by using the mean of the data, we are picking out the maximum likelihood value!
z = (L_obs - mu) / dL
chi2 = np.sum(z ** 2)
dof = N -1
chi2dof = chi2 / dof #this defines the chi^2 per degree of freedom.

print "mean =", mu
print "chi^2 =",chi2dof
print "std = ", np.std(L_obs)
```

```python
# compute the mean and the chi^2/dof
L_obs_2 = L_obs + 0.5 - t ** 2 #change the data so it is not constant with time
mu = np.mean(L_obs_2) #again, this is the MLE for the mean of our model
z = (L_obs_2 - mu) / dL
chi2 = np.sum(z ** 2)
chi2dof = chi2 / dof

print "mean =", mu
print "chi^2 =",chi2dof

#now take a look at the "data" and model
plt.figure()
plt.plot([-0.1, 1.3], [L0, L0], ':k', lw=1) #plot the intirinsic value
plt.errorbar(t, L_obs_2, dL, fmt='.k', ecolor='gray', lw=1) #plot the "data"
plt.plot([-0.1, 1.3], [mu, mu], lw=1, color='red') #plot the intirinsic value
plt.xlim(-0.05, 1.05)
plt.ylim(8.6, 11.4)
plt.xlabel('time')
plt.ylabel('luminosity')
plt.show()
```

## Figure 4.1

```python
# Generate Dataset
np.random.seed(1)
N = 50
L0 = 10
dL = 0.2
t = np.linspace(0, 1, N)
L_obs = np.random.normal(L0, dL, N)

# Plot the results
fig = plt.figure(figsize=(5, 5))
fig.subplots_adjust(left=0.1, right=0.95, wspace=0.05,
                    bottom=0.1, top=0.95, hspace=0.05)
y_vals = [L_obs, L_obs, L_obs, L_obs + 0.5 - t ** 2]
y_errs = [dL, dL * 2, dL / 2, dL]
titles = ['correct errors',
          'overestimated errors',
          'underestimated errors',
          'incorrect model']

for i in range(4):
    ax = fig.add_subplot(2, 2, 1 + i, xticks=[])

    # compute the mean and the chi^2/dof
    mu = np.mean(y_vals[i])
    z = (y_vals[i] - mu) / y_errs[i]
    chi2 = np.sum(z ** 2)
    chi2dof = chi2 / (N - 1)

    # compute the standard deviations of chi^2/dof
    sigma = np.sqrt(2. / (N - 1))
    nsig = (chi2dof - 1) / sigma

    # plot the points with errorbars
    ax.errorbar(t, y_vals[i], y_errs[i], fmt='.k', ecolor='gray', lw=1)
    ax.plot([-0.1, 1.3], [L0, L0], ':k', lw=1)

    # Add labels and text
    ax.text(0.95, 0.95, titles[i], ha='right', va='top',
            transform=ax.transAxes,
            bbox=dict(boxstyle='round', fc='w', ec='k'))
    ax.text(0.02, 0.02, r'$\hat{\mu} = %.2f$' % mu, ha='left', va='bottom',
            transform=ax.transAxes)
    ax.text(0.98, 0.02,
            r'$\chi^2_{\rm dof} = %.2f\, (%.2g\,\sigma)$' % (chi2dof, nsig),
            ha='right', va='bottom', transform=ax.transAxes)

    # set axis limits
    ax.set_xlim(-0.05, 1.05)
    ax.set_ylim(8.6, 11.4)

    # set ticks and labels
    ax.yaxis.set_major_locator(plt.MultipleLocator(1))

    if i > 1:
        ax.set_xlabel('observations')

    if i % 2 == 0:
        ax.set_ylabel('Luminosity')
    else:
        ax.yaxis.set_major_formatter(plt.NullFormatter())

plt.show()
```
